#### Customising Graphs

```python
# Adding Title, X and Y Axis Labels
x = np.arange(1, 5)
plt.plot(x, x)
plt.plot(x, x+1)
plt.plot(x, x+2)

# Title of Graph
plt.title("Title of my Graph")
# Labels
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.show()
```

#### Legend, Grid and Axes
#### LEGEND: for location of legend available options are: 'best', 'upper right', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
#### AXIS: Axis command takes 4 values [xstart, xfinal, ystart, yfinal]

```python
# Adding legend
x = np.arange(1, 5)
plt.plot(x, x)
plt.plot(x, x+1)
plt.plot(x, x+2)

# Title of Graph
plt.title("Line Plot")

# Labels
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Legend
# It takes array as input for each line in plot and assigns the title to it
plt.legend(["One", "Two", "Three"], loc = "best")

plt.show()
```

```python
# Adding legend using 2nd Method
x = np.arange(1, 5)
plt.plot(x, x, label = "One")
plt.plot(x, x+1, label = "Two")
plt.plot(x, x+2, label = "Three")

# Title of Graph
plt.title("Line Plot")

# Labels
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Legend
# It takes array as input for each line in plot and assigns the title to it
# plt.legend(["One", "Two", "Three"], loc = "best")
plt.legend(loc = "lower right")

plt.show()
```

```python
# Grid 
# Enabling Grid and Axis Modification 
x = np.arange(1, 5)
plt.plot(x, x)
plt.plot(x, x+1)
plt.plot(x, x+2)

plt.title("Line plot")

plt.xlabel("X Axis")
plt.xlabel("Y Axis")

plt.legend(["One", "Two", "Three"], loc = "best")

# Grid 
plt.grid(True)

# Axis
plt.axis([0, 7, -2, 10])

plt.show()
```

#### Line and Marker Customisation
#### Available Shapes for Markers are :
#### "s" = square, "p" = pentagon, "." = point, "o" = circle, "d" = diamond, etc.
#### Avalilable Lines are:
#### "-", ":", "--", "-", etc.

```python
x = np.arange(1, 5)
plt.plot(x, x, color = "purple", linestyle = ':', linewidth = 3.0, marker = 'd', markersize = 5.0, markeredgecolor = 'green')
plt.plot(x, x+1, "-s")
plt.plot(x, x+2, "-.p")

plt.title("Line plot")

plt.xlabel("X Axis")
plt.xlabel("Y Axis")

plt.legend(["One", "Two", "Three"], loc = "best")

plt.show()
```

#### Subplot to create multiple plots

```python
# Subplot

# create a figure
fig = plt.figure(figsize = (16, 5), facecolor = 'cyan')

# subplot syntax => fig.add_subplot(rows, columns, position)
rows = 2
columns = 3 

# Follows a grid format => rows x columns 
# Create charts on figure using add_subplot
chart1 = fig.add_subplot(rows, columns, 1)
chart2 = fig.add_subplot(rows, columns, 2)
chart3 = fig.add_subplot(rows, columns, 3)
chart4 = fig.add_subplot(rows, columns, 4)

# Each of the above subplot act as separate plot and can be used to plot anything in space
x = np.arange(1, 5)
chart1.plot(x, x, ":o")
chart2.plot(x, x+1, "-s")
chart3.plot(x, x+2, "-.p")
chart4.plot(x, x+3, ":o", markersize = 10)
# Plot on subplots

plt.show()
```

#### Histogram

```python
# Creating a random array of numbers till 1000
x = np.random.randn(1000)
plt.hist(x)
# plot Histogram on array
plt.show()
```

```python
# Creating a random array
x = np.random.randn(1000)
plt.hist(x, cumulative = True, color = 'purple')
# Plot cummulative histogram on the array
plt.show()
```

```python
# Create a random number generator with a fixed seed for reproducibility
rng = np.random.default_rng(19680801)

N_points = 100000
n_bins = 20

# Generate two normal distributions
dist1 = rng.standard_normal(N_points)
dist2 = 0.4 * rng.standard_normal(N_points) + 5

fig, axs = plt.subplots(1, 2, sharey = True, tight_layout = True)

# We can set the number of bins with the *bins* keyword argument.
axs[0].hist(dist1, bins = n_bins)
axs[1].hist(dist2, bins = n_bins)
```

#### Pie-Chart and Exploded Pie-Chart

```python
plt.figure(figsize = (6, 6))

# labels and values
labels = ["India", "USA","Japan", "Britain", "Germany"]
values = [20, 30, 50, 70, 35]

# plot pie chart without values and with values
plt.pie(values, labels = labels, autopct = "%1.1f%%")

plt.legend(labels)

plt.title("Contries")
plt.show()
```

#### Explode is an array which specifies the fraction of the radius with which to offset each wedge.

```python
plt.figure(figsize = (6, 6))

# labels and values
labels = ["India", "USA","Japan", "Britain", "Germany"]
values = [20, 30, 50, 70, 35]

exp = [0, 0.1, 0.05, 0.02, 0.1]

# plot pie chart without values and with values
plt.pie(values, labels = labels, autopct = "%1.1f%%", explode = exp)

plt.legend(labels)

plt.title("Contries")
plt.show()
```

#### Bar-Charts

```python
plt.figure(figsize = (7, 5), facecolor = "cyan")

# lables and values for bar chart
labels = ["India", "USA", "Japan", "Germany", "UK"]
values = [25, 30, 45, 20, 35]

# Plot the bar chart
plt.bar(labels, values, color = "orange", width = 0.5)

plt.title("Bar Chart")

plt.xlabel("Countries")
plt.ylabel("Values")

plt.show()
```

#### Grouped Bar Charts

```python
plt.figure(figsize = (7, 5), facecolor = "cyan")

# labels and values for grouped chart 
labels = ["India", "USA", "Britaih", "Germany", "Japan"]
values = [25, 30, 20, 10, 35]
values2 = [30, 20, 15, 15, 40]

# Specify Label Positions for tickers on x-axis
positions = np.arange(len(labels))

# Draw 2 bar charts for displaying grouped bars alongside widths
plt.bar(positions, values, color = "orange", width = 0.3)
plt.bar(positions+0.3, values2, color = "blue", width = 0.3)

# give position to display x tickers on X axis, which should be position of bar + half of width of bars
plt.xticks(positions+0.15, labels)

plt.show()
```

#### 3D Scatter Plots

```python
# Give positions for graphs
x = np.random.randn(200)
y = np.random.randn(200)
z = np.random.randn(200)

size = 100 * np.random.randn(200)
colors = np.random.randn(200)

# Create a 3D projection subplot to display our scatter plot
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')

ax.scatter(x, y, z, c = colors, s = size, marker = "^")
plt.show()
```

#### 3D Bar Plots

```python
# Create a 3d projection subplot to display our scatter plot
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection = "3d")
# define distances from plane
x = [1, 2, 3, 4, 5]
y = [3, 6, 2, 8, 5]
# Changing this would result in bar uplifting from plane 
z = [0, 0, 0, 0, 0]

# Length of bar in direction plane
dx = [1, 1, 1, 1, 2]
dy = [1, 1, 2, 1, 1]
dz = [1, 8, 1, 4, 2]

# plot the 3d bar graph using both distances and length of bar
ax.bar3d(x, y, z, dx, dy, dz, color = "gold")

plt.show()
```
