```python
import pandas as pd 
import numpy as np
```

```python
movies = pd.read_csv("Movie-Rating.csv")
```

```python
movies
```

```python
movies.columns
```

```python
movies.columns = ['Film', 'Genre', 'CriticRating', 'AudienceRatings', 'BudgetMillions', 'Year']
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies['Film'] = movies['Film'].astype('category')
movies['Genre'] = movies['Genre'].astype('category')
movies['CriticRating'] = movies['CriticRating'].astype(int)
movies['AudienceRatings'] = movies['AudienceRatings'].astype(int)
movies['BudgetMillions'] = movies['BudgetMillions'].astype(int)
movies['Year'] = movies['Year'].astype('category')
```

```python
movies.info()
```

```python
movies.describe()
```

```python
movies.isnull().sum()
```

```python
import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRatings')
plt.show()
```

```python
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRatings')
plt.show()
```

```python
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRatings', kind = 'hex')
plt.show()
```

```python
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRatings', kind = 'reg')
plt.show()
```

# The value for `kind` must be one of ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid']

```python
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRatings', kind = 'scatter')
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRatings', kind = 'hist')
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRatings', kind = 'hex')
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRatings', kind = 'kde')
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRatings', kind = 'reg')
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRatings', kind = 'resid')
plt.show()
```
